Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.
Matplotlib is a popular plotting library in Python used for creating static, animated, and interactive visualizations. It provides a variety of tools to visualize data in the form of charts and graphs. The Pyplot module is part of Matplotlib and is used for plotting, with a MATLAB-like interface. It is widely used for data analysis and to visualize results in various fields, such as data science, engineering, and scientific computing.

Five plots that can be plotted using the Pyplot module of Matplotlib:

Line plot – Used to visualize data points connected by a line, typically showing trends over time.
Scatter plot – Used to show the relationship between two variables by plotting points on a two-dimensional axis.
Bar plot – Used to display categorical data with rectangular bars representing the data values.
Histogram – Used to represent the distribution of numerical data by dividing it into bins.
Box plot – Used to display the distribution of data based on a five-number summary: minimum, first quartile, median, third quartile, and maximum.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data, plot a scatter plot.

import numpy as np
import matplotlib.pyplot as plt

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Scatter plot
plt.scatter(x, y)

# Title and labels
plt.title("Scatter Plot")
plt.xlabel("X values")
plt.ylabel("Y values")

# Show the plot
plt.show()


Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for plotting
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid for subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title("Line 1")

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title("Line 2")

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title("Line 3")

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title("Line 4")

# Show the plots
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for bar plots
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.figure(figsize=(10, 5))
plt.bar(company, profit)
plt.title("Profit of Companies")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.figure(figsize=(10, 5))
plt.barh(company, profit)
plt.title("Profit of Companies (Horizontal)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create box plot
plt.boxplot([box1, box2], labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
